```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
path = r'C:\Users\chris\OneDrive\Desktop\Cursos\Curso de python para ciencia de datos\Curso_de_Python_para_Ciencia_de_Datos\data\SuperMarket Analysis.csv'
df = pd.read_csv(path, encoding="ISO-8859-1")

# Agrupar y obtener promedio de ventas
avg_sales = df.groupby("Product line")["Sales"].mean()

# Crear gráfico de barras
avg_sales.sort_values().plot(kind="barh", color="skyblue", edgecolor="black")
plt.title("Promedio de ventas por línea de producto")
plt.xlabel("Promedio de ventas")
plt.ylabel("Línea de producto")
plt.tight_layout()
plt.show()
```

```python
# 1. Agrupar y obtener total de ventas por ciudad
ventas_por_ciudad = df.groupby("City")["Sales"].sum().sort_values()

# 2. Crear gráfico de barras horizontales
plt.figure(figsize=(8, 5))  # Opcional, para ajustar tamaño
ventas_por_ciudad.plot(kind="barh", color="red", edgecolor="black")

# 3. Personalización del gráfico
plt.title("Ventas totales por ciudad")
plt.xlabel("Ventas totales")
plt.ylabel("Ciudad")
plt.tight_layout()

# 4. Mostrar gráfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Asumiendo que ya tienes ventas_por_ciudad calculado
sns.set(style="whitegrid")  # Cambia el estilo del gráfico
plt.figure(figsize=(8, 5))
sns.barplot(x=ventas_por_ciudad.values, y=ventas_por_ciudad.index, color='salmon')

plt.title("Ventas totales por ciudad")
plt.xlabel("Ventas totales")
plt.ylabel("Ciudad")
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

ventas_df = ventas_por_ciudad.reset_index()
fig = px.bar(ventas_df, x="Sales", y="City", orientation='h',
             title="Ventas totales por ciudad",
             color="Sales", color_continuous_scale="reds")

fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Suponiendo que ya tienes tu DataFrame 'df'
ventas_por_ciudad = df.groupby("City")["Sales"].sum().reset_index()

# Gráfico con un color diferente por ciudad
fig = px.bar(
    ventas_por_ciudad,
    x="Sales",
    y="City",
    orientation="h",
    color="City",  # esto asigna un color distinto por ciudad automáticamente
    title="Ventas totales por ciudad",
    text="Sales",  # Muestra los valores encima de las barras
    color_discrete_sequence=px.colors.qualitative.Set2  # paleta elegante
)

# Mejoras de estilo
fig.update_layout(
    xaxis_title="Total de Ventas",
    yaxis_title="Ciudad",
    title_font_size=22,
    title_font_color="black",
    template="simple_white"
)

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')

fig.show()

```
