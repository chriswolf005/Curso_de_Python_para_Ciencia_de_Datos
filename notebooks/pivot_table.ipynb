### ðŸ“Œ 1. Tablas dinÃ¡micas (Pivot Tables)

```python
import pandas as pd

# Cargar los datos
path = r'C:\Users\chris\OneDrive\Desktop\Cursos\Curso de python para ciencia de datos\Curso_de_Python_para_Ciencia_de_Datos\data\SuperMarket Analysis.csv'
df = pd.read_csv(path, encoding='ISO-8859-1')
df.head()
```

```python
# Crear una tabla dinÃ¡mica
pivot = pd.pivot_table(df, values="Sales", index="Gender", columns="Product line", aggfunc="mean")
print(pivot)
```

```python
pivot_sum = pd.pivot_table(df, values="Sales", index="Customer type", columns="City", aggfunc="sum")
print(pivot_sum)
```

```python
#Tabla dinÃ¡mica que muestra el promedio precio unitario por linea de producto y metodo de pago
pivot_avg=pd.pivot_table(df,values="Unit price", index="Payment",columns="Product line",aggfunc="mean")
print(pivot_avg)

```

```python
%pip install seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
path = r'C:\Users\chris\OneDrive\Desktop\Cursos\Curso de python para ciencia de datos\Curso_de_Python_para_Ciencia_de_Datos\data\SuperMarket Analysis.csv'
df = pd.read_csv(path, encoding='ISO-8859-1')

# Crear la tabla dinÃ¡mica
pivot_avg = pd.pivot_table(df, values="Unit price", index="Payment", columns="Product line", aggfunc="mean")

# Configurar el grÃ¡fico
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_avg, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("ðŸ’³ Precio Unitario Promedio por Producto y MÃ©todo de Pago", fontsize=14)
plt.xlabel("LÃ­nea de Producto")
plt.ylabel("MÃ©todo de Pago")
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()
```
