```python
import pandas as pd

path=r'C:\Users\chris\OneDrive\Desktop\Cursos\Curso de python para ciencia de datos\Curso_de_Python_para_Ciencia_de_Datos\data\SuperMarket Analysis.csv'
supermarket_data=pd.read_csv(path, encoding='ISO-8859-1')
```

```python
supermarket_data.head()
```

```python
agrupacion_y_promedio_genero=supermarket_data.groupby('Gender')["Sales"].mean()

print(agrupacion_y_promedio_genero)
```

```python
promedio_line=  supermarket_data.groupby("Product line")["Sales"].mean()
print(promedio_line)
```

```python
top_5_productos=promedio_line.nlargest(5)
print("Los 5 productos con mayor promedio de ventas son:")
print(top_5_productos)
worst_5_productos=promedio_line.nsmallest(5)
print("Los 5 productos con menor promedio de ventas son:")
print(worst_5_productos)
```

```python
tipos_de_customer=supermarket_data.groupby("Customer type")["Sales"].mean()
print("Los tipos de clientes son:")
print(tipos_de_customer)
```

### Graficos 📊📈

```python
import matplotlib.pyplot as plt

agrupacion_y_promedio_genero.plot(kind='bar', title='Promedio de Ventas por Género', color=['pink', 'blue'])
plt.ylabel('Promedio de Ventas')
plt.xticks(rotation=0)
plt.show()

```

```python
# Top 5 productos
top_5_productos.plot(kind='bar', title='Top 5 Líneas de Producto con Mayor Promedio de Ventas', color='green')
plt.ylabel('Promedio de Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bottom 5 productos
worst_5_productos.plot(kind='bar', title='Bottom 5 Líneas de Producto con Menor Promedio de Ventas', color='red')
plt.ylabel('Promedio de Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
tipos_de_customer.plot(kind='bar', title='Promedio de Ventas por Tipo de Cliente', color=['green', 'blue'])
plt.ylabel('Promedio de Ventas')
plt.xticks(rotation=0)
plt.show()
```
